```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def clean_train(df):
            
    # Year of Record
    #to comlete the blanks in Year of Record with median
    df['Year of Record'].fillna(df['Year of Record'].median(), inplace = True)
    #to converse the dtype into float64
    df['Year of Record'] = df['Year of Record'].astype('float64')
#     #to normalize
#     s_YoR = df['Year of Record'].min()
#     l_YoR = df['Year of Record'].max()
#     df['Year of Record'] = df['Year of Record'].map(lambda x: (x-s_YoR)/l_YoR)
    
    # Gender
    #to complete the blanks in Gender with "unknown"
    df['Gender'].fillna('unknown', inplace = True)
    #to replace other items with "unknown"
    df['Gender'] = df['Gender'].map(lambda g: 'unknown' if g == '0' else g)
    #to delete all rows containing unknown items
    df.drop(df[df['Gender'] == 'unknown'].index, inplace = True)
    
    # Age
    #to comlete the blanks in Age with median
    df['Age'].fillna(df['Age'].median(), inplace = True)
    #to drop those too high or too low
    df = df.drop(df[df['Age']>100].index)
    #to converse the dtype into float64
    df['Age'] = df['Age'].astype('float64')
#     #to normalize
#     s_A = df['Age'].min()
#     l_A = df['Age'].max()
#     df['Age'] = df['Age'].map(lambda x: (x-s_A)/l_A)
    
    # Size of City
    #to comlete the blanks in Size of City with median
    df['Size of City'].fillna(df['Size of City'].median(), inplace = True)
    #to converse the dtype into float64
    df['Size of City'] = df['Size of City'].astype('float64')
#     #to normalize
#     s_SoC = df['Size of City'].min()
#     l_SoC = df['Size of City'].max()
#     df['Size of City'] = df['Size of City'].map(lambda x: (x-s_SoC)/l_SoC)
    
    # Profession
    #to complete the blanks in Profession with "unknown"
    df['Profession'].fillna('unknown', inplace = True)
    #to delete all rows containing unknown items
    df.drop(df[df['Profession'] == 'unknown'].index, inplace = True)
    
    # University Degree
    #to complete the blanks in University Degree with "unknown"
    df['University Degree'].fillna('unknown', inplace = True)
    #to replace other items with "unknown"
    df['University Degree'] = df['University Degree'].map(lambda u: 'unknown' if u == '0' else u)
    #to delete all rows containing unknown items
    df.drop(df[df['University Degree'] == 'unknown'].index, inplace = True)
    
    # Hair Color & Wears Glasses
    #to delete Hair Color & Wears Glasses
    df.drop(['Hair Color','Wears Glasses'],axis=1,inplace=True)
    
    # Body Height [cm]
    #to comlete the blanks in Body Height [cm] with median
    df['Body Height [cm]'].fillna(df['Body Height [cm]'].median(), inplace = True)
    #to drop those too high or too low
    df = df.drop(df[df['Body Height [cm]']>250].index)
    df = df.drop(df[df['Body Height [cm]']<110].index)
    #to converse the dtype into float64
    df['Body Height [cm]'] = df['Body Height [cm]'].astype('float64')
#     #to normalize
#     s_BH = df['Body Height [cm]'].min()
#     l_BH = df['Body Height [cm]'].max()
#     df['Body Height [cm]'] = df['Body Height [cm]'].map(lambda x: (x-s_BH)/l_BH)
    
    # Income in EUR
    #to drop those too high or too low
    df = df.drop(df[df['Income in EUR']>2000000].index)
    
    return df
```

```python
def clean_test(te):
    
    #Rename
    te.rename(columns={'Income':'Income in EUR'}, inplace=True)
    
    # Year of Record
    #to converse the dtype into float64
    te['Year of Record'] = te['Year of Record'].astype('float64')
    #to comlete the blanks in Year of Record with median
    te['Year of Record'].fillna(te['Year of Record'].median(), inplace = True)
#     #to normalize
#     s_YoR = te['Year of Record'].min()
#     l_YoR = te['Year of Record'].max()
#     te['Year of Record'] = te['Year of Record'].map(lambda x: (x-s_YoR)/l_YoR)
    
    # Gender
    #to complete the blanks in Gender with "unknown"
    te['Gender'].fillna('unknown', inplace = True)
    #to replace other items with "unknown"
    te['Gender'] = te['Gender'].map(lambda g: 'unknown' if g == '0' else g)
    
    # Age
    #to comlete the blanks in Age with median
    te['Age'].fillna(df['Age'].median(), inplace = True)
    #to converse the dtype into float64
    te['Age'] = te['Age'].astype('float64')
#     #to normalize
#     s_A = te['Age'].min()
#     l_A = te['Age'].max()
#     te['Age'] = te['Age'].map(lambda x: (x-s_A)/l_A)
    
    # Size of City
    #to converse the dtype into float64
    te['Size of City'] = te['Size of City'].astype('float64')
#     #to normalize
#     s_SoC = te['Size of City'].min()
#     l_SoC = te['Size of City'].max()
#     te['Size of City'] = te['Size of City'].map(lambda x: (x-s_SoC)/l_SoC)
    
    # Profession
    #to complete the blanks in Profession with "unknown"
    te['Profession'].fillna('unknown', inplace = True)
    
    # University Degree
    #to complete the blanks in University Degree with "unknown"
    te['University Degree'].fillna('unknown', inplace = True)
    #to replace other items with "unknown"
    te['University Degree'] = te['University Degree'].map(lambda u: 'unknown' if u == '0' else u)
    
    # Hair Color & Wears Glasses
    #to delete Hair Color & Wears Glasses
    te.drop(['Hair Color','Wears Glasses'],axis=1,inplace=True)
    
    # Body Height [cm]
    #to comlete the blanks in Body Height [cm] with median
    te['Body Height [cm]'].fillna(te['Body Height [cm]'].median(), inplace = True)
    #to converse the dtype into float64
    te['Body Height [cm]'] = te['Body Height [cm]'].astype('float64')
#     #to normalize
#     s_BH = te['Body Height [cm]'].min()
#     l_BH = te['Body Height [cm]'].max()
#     te['Body Height [cm]'] = te['Body Height [cm]'].map(lambda x: (x-s_BH)/l_BH)
    
    return te
```

```python
df = pd.read_csv('tcd ml 2019-20 income prediction training (with labels).csv', index_col=[0])
te = pd.read_csv('tcd ml 2019-20 income prediction test (without labels).csv', index_col=[0])
df = clean_train(df)
te = clean_test(te)
```

```python
(train_bound, _) = df.shape
```

```python
te.shape
```

```python
#combine test & train
tt_df = pd.concat( [df, te], axis=0)
tt_df
```

```python
# count N/A values
tt_df.shape[0] - tt_df.count()
```

```python
# split Gender
gender = pd.get_dummies(tt_df['Gender'], prefix = 'Gender')
tt_df = tt_df.join(gender)
tt_df.drop(['Gender','Gender_unknown'],axis=1,inplace=True)

tt_df.head()
```

```python
# split Country
country = pd.get_dummies(tt_df['Country'], prefix = 'Country')
tt_df = tt_df.join(country)
tt_df.drop(['Country'],axis=1,inplace=True)

tt_df.head()
```

```python
# split Profession
profession = pd.get_dummies(tt_df['Profession'], prefix = 'Profession')
tt_df = tt_df.join(profession)
tt_df.drop(['Profession','Profession_unknown'],axis=1,inplace=True)

tt_df.head()
```

```python
# split University Degree
university = pd.get_dummies(tt_df['University Degree'], prefix = 'University Degree')
tt_df = tt_df.join(university)
tt_df.drop(['University Degree','University Degree_unknown'],axis=1,inplace=True)

tt_df.head()
```

```python
# seperate train & test
df = tt_df.iloc[:train_bound]
te = tt_df.iloc[train_bound:]
```

```python
# train

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df.drop(['Income in EUR'],axis=1)
y = df['Income in EUR']
```

```python
# train test split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
lin = LinearRegression()
# lin.fit(X_train, y_train)
# lin.score(X_test, y_test)
lin.fit(X, y)
lin.score(X, y)
```

```python
py = lin.predict(te.drop(['Income in EUR'],axis=1))
```
